# Read My Survey Data

# Output

```python
# Age and spending scaterplot
# insight: No corrolation


# Age Distribution.
seaborn.histplot(data['age'], bins = 20, kde=True, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
```

```python
# Sex Distribution.
seaborn.histplot(data['gender'], bins = 2, color='skyblue')
plt.title('Sex Distribution')
plt.xlabel('Sex')
plt.ylabel('# Survey Participants')
```

Gender and Packing Habbit

```python
# Packing Habits by Gender
# BIAS IN DATA. According to this data, female survey participants are more likely to overpack than male participants are.
# However, more women than men filled out the survey.
seaborn.countplot(data=data, x = 'packing_habbit', hue='gender', palette='pastel')
plt.title('Packing Habits')
plt.xlabel('Packing Habit')
plt.ylabel('Count')
```

```python
# Reason for Travel
# Insights: most people are traveling to visit family than any other reason for travel.
seaborn.countplot(y=data['reason'], palette='pastel', order=data['reason'].value_counts().index)
plt.title('Reasons for Travel')
plt.xlabel('Count')
plt.ylabel('Reason')
```

Destination and Hours Spent Packing

```python
# Destination and Hours Spent Packing
column_A = data['desination']
#column_A = data['reason']
column_B = data['packing_hours']

x_axis = column_A
y_axis = column_B

# Destination and Hours Spent Packing
plt.figure(figsize=(12, 6))
plt.bar(x_axis, y_axis, width = 0.4)

plt.ylabel("Type of Destination")
plt.xlabel("Hours Spent Packing for Trip")

plt.title("Hours Spent Packing for Different Destinations")
```

# Goals
- To learn more about the packing habits of infdividuals, such as overpacking, underpacking, or the amount of time people spend packing.
- To find out the biggest reason participants travel.


# Data Set

- __age__: Age of the survey participant.
- __gender__: Gender of the survey participant
- __airport_spending__: The amount of money participants spend on essential travel items while at the airport or final destination in the past year.
- __packing_habbit__: overpacking or underpacking habit of the survey participant
- __travel_frequency__: frequeny
- __airport_spending__: money the survey participant spends on essential travel items while at the airport or at their
- __reason__: the reason for why the survey participant travels
- **destination**: destination of previous trip
- **packing_hours**: The number of hours the survey participant usually spends packing their bags before a trip
- **travel_company**: the people the survey participant travel's with
- **transportation_mode**: the survey participant's common or favorite mode of transportation for travel
luggage
- **previous_destination**: The previous destination of survey participants

# Potential Issues
